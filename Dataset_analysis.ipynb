1. IDZaposlenika: Ovu kolonu smo odabrali kako bismo jedinstveno identificirali svakog radnika u dataset-u. To nam omogućava da pratimo individualne podatke bez miješanja informacija između različitih radnika.

8. Obrazovanje: Specifična oblast obrazovanja može utjecati na radničke interese i zadovoljstvo poslom. Ova kolona nam pomaže da analiziramo da li određene obrazovne pozadine vode ka većem ili manjem napuštanju firme.

10. Uloga: Različite radne uloge dolaze s različitim odgovornostima i izazovima. Ova kolona nam pomaže da analiziramo da li određene uloge imaju veću stopu napuštanja.

15. BrojObuka: Broj obuka može biti indikator podrške i razvoja koje firma pruža svojim radnicima. Više obuka može povećati zadovoljstvo radnika i smanjiti stopu napuštanja.

16. GodineUFirmi: Slična koloni "RadniStaz", ova kolona pokazuje koliko dugo je radnik zaposlen u firmi. Ova informacija je važna za analizu lojalnosti radnika prema firmi.

18. VrijemeSaTrenutnimMenadzerom: Ovaj podatak može pokazati stabilnost menadžmenta i njegov utjecaj na radnike. Česte promjene menadžera mogu negativno utjecati na radničko zadovoljstvo.

19. OcjenaOkoline: Ova kolona predstavlja radničku percepciju radne okoline, što može utjecati na njihovu odluku o ostanku u firmi. Loša radna okolina može povećati stopu napuštanja.

20. OcjenaPosla: Zadovoljstvo poslom je ključni faktor u odlučivanju o ostanku u firmi. Ova kolona omogućava analizu radničkog zadovoljstva i njegov utjecaj na napuštanje firme.

21. OcjenaPreformanse: Ocjena radničke performanse može utjecati na njihovu motivaciju i zadovoljstvo. Niska ocjena performansi može povećati nezadovoljstvo i stopu napuštanja firme.

***2. Kakvi su radni uslovi i starosna dob radnika?***

````python
#Analiza rezultata starosti uposlenih:
age_mean=statistics.mean(starost)
age_moda=statistics.mode(starost)
age_mediana=statistics.median(starost)
age_varijansa=statistics.variance(starost)

#Analiza rezultata udaljenosti od kompanije
dist_mean=statistics.mean(udaljenost_od_kuce)
dist_moda=statistics.mode(udaljenost_od_kuce)
dist_mediana=statistics.median(udaljenost_od_kuce)
dist_varijansa=statistics.variance(udaljenost_od_kuce)

#Analiza rezultata posljednjeg unaprijedjenja:
prom_mean=statistics.mean(zadnje_unaprjedjenje)
prom_moda=statistics.mode(zadnje_unaprjedjenje)
prom_mediana=statistics.median(zadnje_unaprjedjenje)
prom_varijansa=statistics.variance(zadnje_unaprjedjenje)

tabela = PrettyTable(["Kategorija", "Prosjek za kategorije", "Medijana uzorka", "Moda uzorka", "Varijansa"])
tabela.add_row(["Starosna dob", age_mean, age_mediana, age_moda, round(age_varijansa,2)])
tabela.add_row(["Udaljenost od kompanije", dist_mean, dist_mediana, dist_moda, round(dist_varijansa,2)])
tabela.add_row(["Posljednje unaprjeđenje", prom_mean, prom_mediana, prom_moda, round(prom_varijansa,2)])
print(tabela)
````

Možemo vidjeti da su prosjek, medijana i moda, uz blago odstupanje, za starosnu dob  ujednačeni, što znači da kompanije napuštaju ljudi u godinama u kojima su, može se reći, najsposobniji i najproduktivniji. Udaljenost od kompanije očito ne bi trebala predstavljati problem, jer vidimo da je moda uzorka prilično niska, a znamo da prosjek i nije baš najvjerodostojnija informacija. Godine od posljednjeg unaprjeđenja su u prosjeku nešto lošije, ali su ipak medijana i moda prilično stabilni i korektni. Razlog za napuštanje radnika ne možemo pronaći ni u ove tri kategorije.

**UVOD**

Chi-square test zavisnosti je statistički test koji se koristi za ispitivanje veze između dva nominalna ili ordinalna skupa podataka. Ovaj test procjenjuje da li postoji statistički značajna zavisnost između varijabli ili da li su ove varijable nezavisne.

Ovaj test koristi kontingencijsku tablicu za analizu podataka. Kontingencijska tablica (poznata i kao tablica kros-tabulacije, kros-tablica ili dvosmjerna tablica) je raspored u kojem se podaci klasifikuju prema dvjema kategoričkim varijablama.

Osnovna hipoteza (nulta hipoteza) je da nema zavisnosti između varijabli, dok alternativna hipoteza sugeriše postojanje zavisnosti.

Test se izvodi računanjem vrijednosti chi-square statistike, koja mjeri razliku između stvarne raspodjele i očekivane raspodjele. Vrijednost chi-square statistike se zatim upoređuje sa kritičnom vrijednošću iz chi-square raspodjele sa odgovarajućim stepenom slobode. Ako izračunata vrijednost chi-square statistike premašuje kritičnu vrijednost, odbija se nulta hipoteza i zaključuje se da postoji zavisnost između varijabli.

Izabrani parovi varijabli za ispitivanje zavisnosti/nezavisnosti:


*   Bračno stanje i status poslovnog putovanja
*   Status zaposlenja (otpušten/nije otpušten) i spol radnika
*   Područje obrazovanja i odjel

Za nivo signifikantnosti uzimamo α = 0.05.

# ***Kolmogorov–Smirnov i Shapiro–Wilk statistički testovi; QQ-plot kao grafički test***

***UVOD***

***Kolmogorov-Smirnov*** test je moćan statistički alat koji se koristi za procjenu da li određeni skup podataka slijedi pretpostavljenu teorijsku distribuciju. Ovaj test, nazvan po sovjetskim matematičarima Andreju Kolmogorovu i Vladimiru Smirnovu, uspoređuje empirijsku distribuciju podataka s teorijskom distribucijom, pružajući kvantitativnu mjeru odstupanja između njih.

Osnovna ideja Kolmogorov-Smirnov testa je da se izračuna najveća apsolutna razlika između kumulativne distribucije empirijskih podataka (poznate kao empirijska distribucijska funkcija, EDF) i kumulativne distribucijske funkcije (CDF) teorijske distribucije. Ako je ta razlika mala, sugerira se da podaci dobro slijede teorijsku distribuciju. Međutim, ako je razlika velika, to ukazuje na značajno odstupanje od teorijske distribucije.

Jedna od prednosti Kolmogorov-Smirnov testa je njegova neparametarska priroda, što znači da ne pretpostavlja određeni oblik distribucije podataka. Test je posebno koristan u provjerama normalnosti podataka ili u usporedbi dvije različite distribucije. Također, ovaj test može se primijeniti i na diskretne i na kontinuirane distribucije, čineći ga fleksibilnim alatom za različite vrste statističkih analiza.

Zanimljivo je napomenuti da Kolmogorov-Smirnov test ne samo da procjenjuje odstupanje, već također daje p-vrijednost koja pomaže u donošenju zaključaka o statističkoj značajnosti promatranih odstupanja. Što je p-vrijednost manja, to je jači dokaz protiv hipoteze da podaci slijede pretpostavljenu distribuciju. U praksi, ovaj test se široko koristi u mnogim znanstvenim disciplinama, uključujući ekonomiju, biologiju, psihologiju i druge društvene nauke.

# ***Kolmogorov-Smirnov test na varijablu godine***





````python

import pandas as pd
import scipy.stats as stats
import numpy as np

# Učitavanje podataka iz CSV datoteke
tabela = pd.read_csv('/content/drive/MyDrive/datasets/custom_dataset.csv')

# Izdvajanje varijable "Godine"
godine = tabela['Godine']

# Normalizacija varijable "Godine"
godine_normalizovane = (godine - np.mean(godine)) / np.std(godine)

# Izvršavanje Kolmogorov-Smirnov testa na normalizovane podatke
kstest_result = stats.kstest(godine_normalizovane, 'norm')

# Ispis rezultata testa
print("Statistika testa: ", kstest_result.statistic)
print("P-vrijednost: ", kstest_result.pvalue)

# Provjera normalnosti distribucije
alpha = 0.05  # Nivo značajnosti

if kstest_result.pvalue > alpha:
    print("Varijabla 'Godine' prati normalnu distribuciju po Kolmogorov-Smirnovom testu.")
else:
    print("Varijabla 'Godine' ne prati normalnu distribuciju po Kolmogorov-Smirnovom testu.")

````

Na temelju rezultata Kolmogorov-Smirnov testa, cilj nam je utvrditi podudaranje distribucije varijable "Godine" s pretpostavljenom normalnom distribucijom. Rezultati testa pokazuju da nema statistički značajne razlike između empirijske distribucije (distribucije varijable "Godine") i teoretske normalne distribucije. P-vrijednost iznosi 0.089, što je iznad uobičajenog praga značajnosti od 0.05. To ukazuje na to da nema dovoljno dokaza za odbacivanje hipoteze o normalnosti distribucije.

Visoka vrijednost test statistike nije prisutna, što dodatno potvrđuje da varijabla "Godine" ne odstupa značajno od normalne distribucije. Ovaj rezultat sugerira da podaci o godinama zaposlenika prate obrazac normalne distribucije, što je povoljno za daljnje statističke analize koje se oslanjaju na pretpostavku normalnosti podataka.

Iako je uvijek važno razmotriti moguće prisutne outliere, asimetriju ili izraženu kurtozu u podacima, rezultati Kolmogorov-Smirnov testa pružaju dovoljno osnove za zaključak da varijabla "Godine" prati normalnu distribuciju. Stoga možemo nastaviti s korištenjem metoda i testova koji zahtijevaju normalnost distribucije, bez potrebe za dodatnim transformacijama podataka.

# ***Kolmogorov-Smirnov test za varijablu "UdaljenostOdDoma"***

````python
import pandas as pd
import scipy.stats as stats

# Učitavanje podataka iz CSV datoteke
tabela = pd.read_csv('/content/drive/MyDrive/datasets/custom_dataset.csv')

# Izdvajanje varijable "Udaljenost od doma"
udaljenost_od_doma = tabela['UdaljenostOdDoma']

# Standardizacija varijable
mean = udaljenost_od_doma.mean()
std = udaljenost_od_doma.std()
standardized_data = (udaljenost_od_doma - mean) / std

# Izvršavanje Kolmogorov-Smirnov testa
kstest_result = stats.kstest(standardized_data, 'norm')

# Ispis rezultata testa
print("Statistika testa:", kstest_result.statistic)
print("P-vrijednost:", kstest_result.pvalue)

# Provjera normalnosti distribucije
alpha = 0.05  # Nivo značajnosti

if kstest_result.pvalue > alpha:
    print("Varijabla 'Udaljenost od doma' prati normalnu distribuciju po Kolmogorov-Smirnovom testu.")
else:
    print("Varijabla 'Udaljenost od doma' ne prati normalnu distribuciju po Kolmogorov-Smirnovom testu.")
````

# ***Zaključak i pojašnjenje***

Iz rezultata Kolmogorov-Smirnovog i Shapiro-Wilkovog testa jasno je da većina varijabli ne prate normalnu distribuciju. Ova odstupanja mogu biti uzrokovana različitim faktorima, koji su često povezani sa specifičnim karakteristikama distribucije podataka. Razmotrimo detaljno moguće razloge za ovu pojavu:



Asimetrija (Skewedness): Ako distribucija podataka nije simetrična, tj. ako pokazuje izrazitu nagnutost u jednom smjeru, rezultira odstupanjem od normalne distribucije koja je simetrična. Na primjer, kada većina vrijednosti varijable leži blizu nižih vrijednosti, dok se rep distribucije proteže prema višim vrijednostima, govorimo o pozitivnoj asimetriji. Suprotno tome, negativna asimetrija se javlja kada su niže vrijednosti koncentrirane u repu distribucije. Asimetrična distribucija može značajno utjecati na rezultate testova normalnosti.



Izvanredne vrijednosti (Outliers): Prisutnost ekstremnih vrijednosti može narušiti pretpostavke normalne distribucije. Ove izvanredne vrijednosti, koje su značajno udaljene od većine podataka, mogu izobličiti distribuciju, čineći je daleko od normalne. Outlieri mogu nastati zbog različitih razloga, uključujući pogreške pri unosu podataka, specifične događaje ili prirodne varijacije u podacima.



Multi-modalnost: Ako distribucija podataka ima više vrhova, tj. ako se podaci grupiraju u nekoliko jasno različitih skupina, govorimo o multi-modalnosti. Takva distribucija može ukazivati na prisutnost više subpopulacija unutar skupa podataka. Na primjer, u istraživanjima zadovoljstva zaposlenika, različiti odjeli mogu imati različite distribucije zadovoljstva, što rezultira višestrukim vrhovima u ukupnoj distribuciji.



Transformacije podataka: Ponekad, primjena matematičkih transformacija kao što su logaritamske, kvadratne ili recipročnih transformacija može pomoći u približavanju distribucije normalnoj. Transformacije podataka mogu smanjiti asimetriju i utjecaj outliera, čineći distribuciju prikladnijom za analizu koja pretpostavlja normalnost.



Sezonski ili ciklični efekti: U nekim situacijama, distribucija podataka može biti pod utjecajem sezonskih ili cikličnih varijacija. Na primjer, prodaja proizvoda može imati sezonske vrhove tokom praznika, što dovodi do odstupanja od normalne distribucije u mjesečnim podacima.



Prirodna varijabilnost: U mnogim prirodnim i društvenim fenomenima, varijabilnost podataka može biti inherentna karakteristika koja ne slijedi normalnu distribuciju. Primjeri uključuju biološke mjere poput krvnog pritiska ili nivoa glukoze u krvi, koje često pokazuju prirodnu varijabilnost izvan normalnih granica.



Učinci mješanja distribucija: Kada se podaci prikupljaju iz različitih izvora ili populacija, može doći do miješanja različitih distribucija. Ovo može dovesti do agregatne distribucije koja ne slijedi normalnu distribuciju. Primjeri uključuju ankete koje obuhvaćaju različite demografske grupe ili tržišta.



Veličina uzorka: Shapiro-Wilkov test je posebno osjetljiv na veličinu uzorka. Za vrlo male uzorke, test može biti manje pouzdan, dok kod velikih uzoraka čak i mala odstupanja od normalnosti mogu rezultirati značajnim p-vrijednostima. Stoga je važno uzeti u obzir veličinu uzorka prilikom interpretacije rezultata.



U zaključku, rezultati Kolmogorov-Smirnovog i Shapiro-Wilkovog testa naglašavaju važnost razumijevanja karakteristika podataka prije primjene statističkih metoda koje pretpostavljaju normalnost. Kombinacija grafičkih prikaza (poput QQ-plota i histograma) s formalnim testovima normalnosti pruža sveobuhvatan pristup procjeni distribucije podataka. Daljnja statistička analiza i transformacije podataka mogu pomoći u prilagođavanju distribucije, omogućujući preciznije i pouzdanije zaključke u istraživačkim studijama.


# Estimacija intervala pouzdanosti srednje vrijednosti za numeričke varijable

*UVOD*

Estimacija nekog parametra populacije, kao što su to recimo srednja vrijednost, proporcija i varijansa, predstavljaju ključni dio statičke inferencije. Point estimacija je procjena koja koristi vrijednost parametra iz uzorka kao aproksimaciju nepoznatog parametra populacije. Međutim, često se preferira da se odredi interval unutar kojeg očekujemo da pronađemo vrijednost parametra, u našem slučaju srednju vrijednost. Numeričke varijable za koje ćemo uraditi estimaciju intervala pouzdanosti srednje vrijednosti su:

8.   Ocjena okoline
9.   Ocjena posla
10.  Ocjena performanse

Sada je potrebno odabrati metod, odnosno pristup za izračunavanje intervala pouzdanosti srednje vrijednosti. Kako je rečeno, to ćemo uraditi na osnovu veličine uzorka.

````python
numericke_kolone = tabela.select_dtypes(include=[np.number])

vel_uzorka = numericke_kolone.apply(lambda x: x.dropna().shape[0])

vel_uzorka_df = vel_uzorka.to_frame(name="Velicina uzorka")

print(vel_uzorka_df)
````

Kako imamo veličinu uzorka veću od 30, interval pouzdanosti možemo izračunati i pomoću studentove **t-raspodjele** i **normalne distribucije**. Ovo nam omogućava *Centralni granični teorem* koji kaže da distribucija uzorka srednje vrijednosti biti uvijek normalno distribuirana, sve dok je veličina uzorka dovoljno velika. Za dovoljno veliku veličinu uzorka smatra se uzorak n ≥ 30.

**STUDENTOVA T-RASPODJELA**

U slučaju kada nam je varijansa populacije nepoznata, možemo koristiti t-raspodjelu. Prije samog računanja intervala pouzdanosti, treba naglasiti da što je interval pouzdanosti širi, možemo biti pouzdaniji da interval sadrži vrijednost parametra koji estimiramo. Ipak, bolje nam je da imamo kraći interval sa 95% pouzdanosti, nego značajno širi interval sa 99% pouzdanosti. To ćemo i pokazati tako što ćemo uzeti nivo pouzdanosti prvo 95%, a zatim 99%, što znači da će koeficijent pouzdanosti α uzimati vrijednosti 0.05 i 0.01 redom.



```
alpha = 0.05

nivo_pouzdanosti = (1 - alpha) * 100
```



````python
numericke_kolone = tabela.select_dtypes(include=[np.number]).drop(columns=["IDZaposlenika","StandardniSati","OcjenaPosla","RadniStaz","OcjenaOkoline"])

def interval_pouzdanosti_t(column, pouzdanost=0.95):
    n = len(column.dropna())
    srednja_vrijednost = np.mean(column)
    std_dev = np.std(column, ddof=1)
    sem = std_dev / np.sqrt(n)
    t_value = stats.t.ppf((1 + pouzdanost) / 2., df=n-1)
    donja_granica = srednja_vrijednost - t_value * sem
    gornja_granica = srednja_vrijednost + t_value * sem
    return srednja_vrijednost, donja_granica, gornja_granica

interval_pouzdanosti_95 = {}
for col in numericke_kolone:
    interval_pouzdanosti_95[col] = interval_pouzdanosti_t(tabela[col])

interval_pouzdanosti_df = pd.DataFrame(interval_pouzdanosti_95, index=['Srednja vrijednost', 'Donja granica', 'Gornja granica']).T

print(interval_pouzdanosti_df)
````

Kao što možemo da vidimo, podaci o intervalima pouzdanosti srednje vrijednosti za numeričke varijable prikazani su tabelarno. **NORMALNA DISTRIBUCIJA**

Interval pouzdanosti srednje vrijednosti moguće je izračunati korištenjem ove metode zbog dovoljno velike veličine uzorka. Inače ova metoda se može primjeniti kada uzorak dolazi iz normalne populacije, ili ukoluko je n (veličina uzorka) dovoljno veliko. Pogodna veličina uzorka nam omogućava korištenje ove metode, budući da smo već naveli da varijable uglavnom ne prate normalnu distribuciju. Također treba napomenuti da  bi trebali dobiti isti rezultat kao kada smo primjenili metod t-raspodjele. Zbog toga ćemo koristiti iste nivoe pouzdanosti. Dakle intervale pouzdanosti srednje vrijednosti za numeričke varijable ćemo ispitati, uzimajući vrijednost α da je 0.05, te zatim 0.01.

````python
def interval_pouzdanosti95(tabela, pouzdanost=0.95):
    n = len(tabela)
    srednja_vrijednost_95 = np.mean(tabela)
    standardna_greska = stats.sem(tabela)
    h = standardna_greska * stats.norm.ppf((1 + pouzdanost) / 2.)
    donja_granica_95=srednja_vrijednost_95 - h
    gornja_granica_95=srednja_vrijednost_95 + h
    return srednja_vrijednost_95, donja_granica_95, gornja_granica_95

intervali_pouzdanosti = {}
for col in numericke_kolone:
    if col in tabela.columns:
        podatak = tabela[col].dropna().astype(float)
        srednja_vrijednost_95, donja_granica_95, gornja_granica_95 = interval_pouzdanosti95(podatak)
        intervali_pouzdanosti[col] = {
            'Srednja vrijednost': srednja_vrijednost_95,
            'Donja granica': donja_granica_95,
            'Gornja_granica': gornja_granica_95
        }

intervali_pouzdanosti_df = pd.DataFrame(intervali_pouzdanosti).T

print(intervali_pouzdanosti_df)
````

````python
def interval_pouzdanosti99(tabela, pouzdanost=0.99):
    n = len(tabela)
    srednja_vrijednost_99 = np.mean(tabela)
    standardna_greska = stats.sem(tabela)
    h = standardna_greska * stats.norm.ppf((1 + pouzdanost) / 2)
    donja_granica_99=srednja_vrijednost_99 - h
    gornja_granica_99=srednja_vrijednost_99 + h
    return srednja_vrijednost_99, donja_granica_99, gornja_granica_99

intervali_pouzdanosti = {}
for col in numericke_kolone:
    if col in tabela.columns:
        podatak = tabela[col].dropna().astype(float)
        srednja_vrijednost_99, donja_granica_99, gornja_granica_99 = interval_pouzdanosti99(podatak)
        intervali_pouzdanosti[col] = {
            'Srednja vrijednost': srednja_vrijednost_99,
            'Donja granica': donja_granica_99,
            'Gornja_granica': gornja_granica_99
        }


intervali_pouzdanosti_df = pd.DataFrame(intervali_pouzdanosti).T

print(intervali_pouzdanosti_df)
````

Prije analize svakog od datih numeričkih podataka, upravo u svrhu što detaljnije analize, bilo bi zgodno još proračunati standardnu grešku estimatora i standardnu devijaciju uzorka za sve numeričke varijable osim za one za koje smo rekli da ćemo ih isključiti iz daljeg proračuna.Standardna devijacija nam govori koliko su podaci u uzorku rašireni oko prosjeka. Što je veća devijacija, to je veća raznolikost među podacima. Standardna greška nam govori koliko procjenjena srednje vrijednosti uzorka odstupa od stvarne srednje vrijednosti populacije. Manja standardna greška ukazuje na veću preciznost procjene srednje vrijednosti.

````python
standardne_greske = {}
standardne_devijacije = {}

for col in numericke_kolone:
    if col in tabela.columns:
        podatak = tabela[col].dropna().astype(float)
        standardne_greske[col] = stats.sem(podatak)
        standardne_devijacije[col] = np.std(podatak, ddof=1)

standardne_greske_df = pd.DataFrame(standardne_greske, index=['Standardna greška']).T
standardne_devijacije_df = pd.DataFrame(standardne_devijacije, index=['Standardna devijacija']).T

print("\nStandardne greške estimatora:")
print(standardne_greske_df)
print("\nStandardne devijacije:")
print(standardne_devijacije_df)
````

# Postavljanje i statičko ispitivanje hipoteza

**HIPOTEZA I**

Prva hipoteza će biti vezana za varijablu GodineUFirmi. Potrebno je postaviti nultu i alternativnu hipotezu. Nulta hipoteza je osnovna pretpostavka koja se testira i obično predstavlja stanje "nema efekta" ili "nema razlike". Ona je formulirana na način da može biti jasno opovrgnuta ili odbijena. Alternativna hipoteza je suprotna nultoj hipotezi i predstavlja tvrdnju koju istraživač želi dokazati. U skladu s tim, za navedenu varijablu one će glasiti:

H0: Prosječan broj godina provedenih u firmi je manji 5 godina.

H1: Prosječan broj godina provedenih u firmi je veći od 5 godina

Mi sada želimo da testiramo da li je srednja vrijednost varijable iz uzorka veća ili manja od hipotetičke vrijednosti (𝜇=5). To ćemo uraditi jednostranim-t testom. Tako ćemo ispitati da li postoji statistički značajna razlika između srednje vrijednosti varijable i poznate srednje vrijednosti. Kada sprovedemo t-test, dobićemo p-vrijednost koja će nam pomoći da odlučimo da li da odbacimo nultu hipotezu ili ne. Ona konkretno predstavlja vjerovatnoću da dobijemo rezultat jednak ili ekstremniji od onoga kojeg smo vidjeli, pod pretpostavkom da je nulta hipoteza tačna. P-vrijednost se poredi sa odabranim nivom značajnosti, koji se obično uzima da je 0.05. Mi uzimamo da je rezultat statistički značajan ukoliko on označava da je razlika ili odnos između varijabli dovoljno veliki da se može smatrati stvarnim, a ne slučajnošću. Kada je rezultat statistički značajan znači da imamo dovoljno dokaza da odbacimo nultu hipotezu i prihvatimo alternativnu.

````python

godine_u_firmi = tabela['GodineUFirmi']

test_statistika, p_vrijednost = stats.ttest_1samp(godine_u_firmi, popmean=5)

alpha = 0.05
if p_vrijednost < alpha:
    print("Rezultati su statistički značajni. Prosječan broj godina provedenih u kompaniji je veći od 5 godina.")
else:
    print("Rezultati nisu statistički značajni. Prosječan broj godina provedenih u kompaniji je manji od 5 godina.")
````

Na osnovu sprovedenog jednostranog t-testa, prosečan broj godina provedenih u firmi je statistički značajno veći od 5 godina (p < 0.05). To sugeriše da zaposleni imaju tendenciju da ostaju duže u firmi nego što smo inicijalno pretpostavili.

H0: Prosječan ocjena okoline je manja 3.

````python
ocjena_okoline = tabela['OcjenaOkoline']
test_statistika_okoline, p_vrijednost_okoline = stats.ttest_1samp(ocjena_okoline, popmean=3)

alpha = 0.05
if p_vrijednost < alpha:
    print("Rezultati su statistički značajni. Prosječan ocjena okoline veća od 3 .")
else:
    print("Rezultati nisu statistički značajni. Prosječan ocjena okoline je manja od 3.")
````

Rezultat nije statistički značajan, dakle ne možemo podržati alternativnu hipotezu. Pošto je p-vrijednost veća od nivoa značajnosti 0.05, ne možemo odbaciti nultu hipotezu. To znači da nema dovoljno dokaza da zaključimo da je prosječna ocena okoline veća od 3. Rezultati sugerišu da zaposleni ocenjuju okolinu u kojoj rade kao prosječnu ili neutralnu

**HIPOTEZA IV**

H0: Ne postoji značajna korelacija između broja obuka koje zaposlenici pohađaju i njihove ocjene posla.

H1: Postoji značajna korelacija između broja obuka koje zaposlenici pohađaju i njihove ocjene posla.

Sada trebamo sprovesti Pearsonov korealciju kako bi dobili Pearsonov korelacijski koeficijent. To je mjera koja kvantifikuje smjer linearne veze između dvije varijable. Vrijednost 1 predstavlja pozitivnu korelaciju, vrijednost -1 predstavlja negativnu korealciju, te vrijednost 0 predstavlja da nema korelacije. Ako je vrijednost između 0.1 i 0.3 to je slaba korelacija, od 0.3 do 0.5 je umjerena i iznad 0.5 je snažna korelacija. Da bi sproveli postupak moramo ukloniti vrijednosti koje su NaN.

````python
data_clean = tabela.dropna(subset=['BrojObuka', 'OcjenaPosla'])

broj_obuka_clean = data_clean['BrojObuka']
ocjena_posla_clean = data_clean['OcjenaPosla']

korrelacija_obuka_posla, p_vrijednost_korrelacija = stats.pearsonr(broj_obuka_clean, ocjena_posla_clean)

alpha=0.05
if p_vrijednost < alpha:
    print("Korelacija je statistički značajna.")
else:
    print("Korelacija nije statistički značajna.")
````

Vrijednost Pearsonovog korelacijskog koficijenta je 0.04, što ukazuje na veoma slabu korelaciju između broja obuka i ocjene posla. Pošto je p-vrijednost 0.6210 veća od nivoa značajnosti od 0.05, ne možemo odbaciti nultu hipotezu. Ovo sugeriše da broj obuka koje zaposlenici pohađaju nije povezan sa njihovom ocjenom posla. Drugim riječima, broj obuka vjerovatno nema utjecaj na to kako zaposlenici ocjenjuju svoj posao.

````python
plt.figure(figsize=(10, 6))
sns.regplot(x=broj_obuka_clean, y=ocjena_posla_clean, ci=None, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.xlabel('Broj Obuka')
plt.ylabel('Ocjena Posla')
plt.title('Korelacija između Broja Obuka i Ocjene Posla')
plt.grid(True)
plt.show()
````

Svaka tačka na scatter plot-u predstavlja par podataka za dvije varijable. Gledajući kako su tačke raspoređene, možemo proceniti da li postoji određeni obrazac ili trend. Regresiona linija prikazuje pravolinijski trend odnosa između dve varijable. Ako su tačke blizu regresione linije, to znači da postoji jaka linearna veza između varijabli. #Bonus zadatak: Model linearne regresije



````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

tabela['RadniStaz'] = tabela['RadniStaz'].fillna(tabela['RadniStaz'].median())
tabela['OcjenaOkoline'] = tabela['OcjenaOkoline'].fillna(tabela['OcjenaOkoline'].median())
tabela['OcjenaPosla'] = tabela['OcjenaPosla'].fillna(tabela['OcjenaPosla'].median())

selected_features = ['Godine', 'MjesecnaPlata', 'RadniStaz', 'OcjenaOkoline', 'OcjenaPosla', 'VrijemeOdZadnjegUnaprijedenja', 'VrijemeSaTrenutnimMenadzerom']
X = tabela[selected_features]
y = tabela['GodineUFirmi']

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)

y_pred = ridge.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error:', mse)
print('R² Score:', r2)

plt.figure(figsize=(10, 6))
sc = plt.scatter(y_test, y_pred, c=y_pred, cmap='viridis', alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--r', linewidth=2)
plt.xlabel('Stvarne vrijednosti')
plt.ylabel('Vrijednosti dobivene modelom')
plt.title('Poređenje vrijednosti')
plt.show()
````


Na temelju rezultata modela s niskim Mean Squared Error (MSE) i dobrim R² score, možemo zaključiti da postoji određena veza između ulaznih parametara ('Godine', 'MjesecnaPlata', 'RadniStaz', 'OcjenaOkoline', 'OcjenaPosla', 'VrijemeOdZadnjegUnaprijedenja', 'VrijemeSaTrenutnimMenadzerom') i ciljne varijable ('GodineUFirmi').

Raspon naše ciljne varijable, godine u firmi, iznosi 35. Vrijednost MSE je razumna u odnosu na raspon ciljne varijable, što ukazuje na relativno dobre performanse modela.
